```python
width=0.5
plt.barh(list(fb["Stadium"].value_counts()[0:5].keys()),list(fb["Stadium"].value_counts()[0:5]),width)
plt.title("Maximum Match in Stadium",fontsize=20)
plt.ylabel("Stadium Name",color="red")
plt.xlabel("No. of Matches",color="red")

plt.savefig("Max_M_SDM.jpg",dpi=1000,facecolor="y",bbox_inches="tight",transparent=True)
plt.show()
```

```python
h_goal=fb[["Home Team Goals","Home Team Name"]]
```

```python
h_goal=h_goal.sort_values(by="Home Team Goals",ascending=False)
h_goal.head(15)
```

```python
s=[250]
plt.scatter(list(h_goal["Home Team Name"][0:10]),list(h_goal["Home Team Goals"][0:10]),edgecolor="y",
            linewidth=2,sizes=s,marker="*")
plt.title("Highest Goals by Home Teams",fontsize=20,color="red")
plt.xlabel("Team Name",fontsize=15)
plt.ylabel("goals",fontsize=15)

plt.savefig("H_Goal_Home.jpg",dpi=1000,facecolor="orange")
plt.show()
```

```python
a_goal=fb[["Away Team Goals","Away Team Name"]]
a_goal
```

```python
a_goal=a_goal.sort_values(["Away Team Goals"],ascending=False)
a_goal
```

```python
a_goal.head(15)
```

```python
colors=[10,22,37,19,48,4]
s=[200,250,300,325,350,375]
plt.scatter(list(a_goal["Away Team Name"][0:6]),list(a_goal["Away Team Goals"][0:6]),sizes=s,
            marker="*",c=colors,cmap="viridis")
l=plt.colorbar()
l.set_label("Color Bar",fontsize=25,color="g")
plt.title("Highest goal by Away Teams",fontsize=20,color="red")
plt.xlabel("Team Name",fontsize=15,color="orange")
plt.ylabel("No. of Goals",fontsize=15,color="orange")

plt.savefig("H_Goal_Away.jpg",dpi=1000,facecolor="pink")
plt.show()
```

```python
plt.hist(list(fb["Attendance"]),color="y",label="Attendence",edgecolor="green")
plt.axvline(100000,color="red")
plt.legend()
plt.title("Total Attendance in All Matches",fontsize=25,color="red")

plt.savefig("Hist_Attandence",dpi=1000,facecolor="pink")
plt.show()
```

```python
top_att=fb[["Datetime","Attendance"]].sort_values(by=["Attendance"],ascending=False).head(4)
top_att
```

```python
plt.step(list(top_att["Datetime"]),list(top_att["Attendance"]),marker="*",mfc="r",ms=10)
plt.title("Highest Attendance",fontsize=20,color="red")
plt.xlabel("Date and Time",fontsize=15,color="m")
plt.grid()

plt.savefig("H_Attandance",dpi=1000)
plt.show()
```

```python
low_att=fb[["Datetime","Attendance"]].sort_values(by=["Attendance"],ascending=False).tail(4)
low_att
```

```python
plt.plot(list(low_att["Datetime"]),list(low_att["Attendance"]),marker="*",mfc="g",ms=20)
plt.title("Lowest Attendance",fontsize=25,color="r")
plt.xlabel("Date and Time",fontsize=15,color="orange")
plt.grid()

plt.savefig("L_Attandance",dpi=1000)
plt.show()
```
